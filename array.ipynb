array berisi nilai mahasiswa

```python
import numpy as np
print("\n2. Array berisi nilai mahasiswa dan tampilkan nilai tertinggi:  ")
nilai_mahasiswa = np.array([85, 90, 78, 92, 88, 95, 80, 91, 89, 94, 87, 93, 86, 90, 91])
nilai_tertinggi = np.max(nilai_mahasiswa)
print(nilai_mahasiswa)
print("Nilai tertinggi:", nilai_tertinggi)
```

```python
nilai_terendah = np.min(nilai_mahasiswa)
print("Nilai terendah:", nilai_terendah)
```

```python
nilai_rata = np.mean(nilai_mahasiswa)
print("nilai_rata: ", nilai_rata)
```

```python
import matplotlib.pyplot as plt

jumlah_mahasiswa = len(nilai_mahasiswa) # Define jumlah_mahasiswa
labels = ['Total Mahasiswa'] # Adjust label for a single slice pie chart
sizes = [jumlah_mahasiswa] # sizes expects an array-like object, wrap jumlah_mahasiswa in a list

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Grafik Pie Chart Jumlah Anggota")
plt.axis('equal')  # Membuat grafik menjadi lingkaran
plt.show()
```
