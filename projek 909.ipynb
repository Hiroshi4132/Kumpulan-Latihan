```python
# definisikan list kosong untuk tempat menyimpan angka
my_list = []
```

```python
#input 10 angka
for i in range(10):
  angka = int (input("masukan angka :"))
  my_list.append(angka)
```

```python
print("List:", my_list)
print("Nilai maksimum:", max(my_list))
print("Nilai minimum:", min(my_list))
print("Rata-rata:", sum(my_list)/len(my_list))
```

```python
#Inisialisasi penghitung
jumlah_lulus = 0
jumlah_tidak_lulus = 0

for nilai in my_list:
    if nilai >= 60:
        jumlah_lulus += 1
    else:
        jumlah_tidak_lulus += 1

print(f"Jumlah siswa yang lulus: {jumlah_lulus}")
print(f"Jumlah siswa yang tidak lulus: {jumlah_tidak_lulus}")
```

```python
import matplotlib.pyplot as plt
plt.bar(range(len(my_list)), my_list)
plt.title("Grafik Array")
plt.xlabel("Indeks")
plt.ylabel("Nilai")
plt.show()
```

```python
import matplotlib.pyplot as plt

kategori = ['Lulus', 'Tidak Lulus']
jumlah = [jumlah_lulus, jumlah_tidak_lulus]

plt.figure(figsize=(8, 5))
plt.bar(kategori, jumlah, color=['#4CAF50', '#F44336']) # Hijau untuk lulus, Merah untuk tidak

plt.title('Statistik Kelulusan Mahasiswa')
plt.xlabel('Status')
plt.ylabel('Jumlah Mahasiswa')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
